```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
#X_train,X_test,y_train,y_test=train_test_split(df,test_size=0.2,random_state=42)
#Split X and y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.svm import SVC
```

```python
svm_Default=SVC()
svm_Default.fit(X_train,y_train)
y_pred_SVM_Default=svm_Default.predict(X_test)
```

```python
SVM_Default_Accuracy=100.0* accuracy_score(y_test,y_pred_SVM_Default)
print("SVM Default Kernel Accuracy:" ,str(SVM_Default_Accuracy))
```

```python
svm_Linear=SVC(kernel='linear',C=1.0 ,random_state=42)
svm_Linear.fit(X_train,y_train)
y_pred_SVM_Linear=svm_Linear.predict(X_test)
```

```python
SVM_Linear_Accuracy=100.0* accuracy_score(y_test,y_pred_SVM_Linear)
print("SVM Default Kernel Accuracy:" ,str(SVM_Linear_Accuracy))
```

```python
svm_Poly=SVC(kernel='poly',degree=1.0,gamma=1)
svm_Poly.fit(X_train,y_train)
y_pred_SVM_Poly=svm_Poly.predict(X_test)
```

```python
SVM_Poly_Accuracy = 100.00* accuracy_score(y_test,y_pred_SVM_Poly)
print("SVM Poly Kernel Accuracy:" ,str(SVM_Poly_Accuracy))
```

```python
svm_Sigmoid=SVC(kernel='sigmoid',degree=1.0,random_state=42)
svm_Sigmoid.fit(X_train,y_train)
y_pred_SVM_Sigmoid=svm_Sigmoid.predict(X_test)
```

```python
SVM_Sigmoid_Accuracy = 100.00*accuracy_score(y_test,y_pred_SVM_Sigmoid)
print("SVM Poly Kernel Accuracy:" ,str(SVM_Sigmoid_Accuracy))
```
